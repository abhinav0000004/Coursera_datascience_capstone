```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])
   
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=df_toronto['Neighborhood'],
                                latitudes=df_toronto['Latitude'],
                                longitudes=df_toronto['Longitude'])
toronto_venues.shape
```

```python
toronto_venues.groupby('Neighborhood').count().head()
```

```python
"Indian Restaurant" in toronto_venues['Venue Category'].unique()
```

```python
tab = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
tab['Neighborhoods'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [tab.columns[-1]] + list(tab.columns[:-1])
tab = tab[fixed_columns]

print(tab.shape)
tab.head()
```

```python
grp = tab.groupby(["Neighborhoods"]).mean().reset_index()

print(grp.shape)
grp.head()
```

```python
india = grp[["Neighborhoods","Indian Restaurant"]]
```

```python
india.head(9)
```

```python
from sklearn.cluster import KMeans
toclusters = 4

clus = india.drop(["Neighborhoods"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=toclusters, random_state=1)
kmeans.fit_transform(clus)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:20]
```

```python
Final = india.copy()

# add clustering labels
Final["Cluster Labels"] = kmeans.labels_
```

```python
Final.rename(columns={"Neighborhoods": "Neighborhood"}, inplace=True)
Final.head(5)
```

```python
Final = Final.join(toronto_venues.set_index("Neighborhood"), on="Neighborhood")

print(Final.shape)
Final.head()
```

```python
Final.sort_values(["Cluster Labels"], inplace=True)
```

```python
map_clusters = folium.Map(location=[lat_toronto, lon_toronto],zoom_start=14)
# add markers to the map
markers_colors={}
markers_colors[0] = 'green'
markers_colors[1] = 'black'
markers_colors[2] = 'blue'
markers_colors[3] = 'yellow'
markers_colors[4] = 'pink'
markers_colors[5] = 'brown'
for lat, lon, cluster in zip(Final['Neighborhood Latitude'], Final['Neighborhood Longitude'], Final['Cluster Labels']):
    
    
    folium.CircleMarker(
        [lat, lon],
        radius=5,
       
        color =markers_colors[cluster],
        fill_color=markers_colors[cluster],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
